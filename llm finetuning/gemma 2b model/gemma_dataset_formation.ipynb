```python
import pandas as pd
with open("/content/yungridataset.txt", 'r') as file:

    file_contents = file.readlines()

ques=[]
ans=[]
for line in file_contents:
   if line !="\n":
      if line.split()[0]=="Question:":
          ques.append(line.replace("\n",""))
      else:
          ans.append(line.replace("\n",""))

# remove Question: ques
f_que=[]
for i in ques:
  f_que.append(i.replace("Question:",""))

# remove Question: ques
f_ans=[]
for i in ans:
  f_ans.append(i.replace("Answer:",""))

formatted_dataset = []

for question, answer in zip(f_que, f_ans):
    formatted_dataset.append(f"<bos><start_of_turn>user\nQuestion: {question}\n<end_of_turn>\n<start_of_turn>model\nAnswer: {answer}\n<end_of_turn>")

formatted_dataset[:5]

```

```python
import pandas as pd

df=pd.DataFrame(formatted_dataset)
df.head()
```

```python
df.to_csv('yungrioutput.csv', index=False)
```
