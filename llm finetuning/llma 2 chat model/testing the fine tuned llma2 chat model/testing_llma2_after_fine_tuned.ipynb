```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

quantization_config = BitsAndBytesConfig(load_in_8bit=True)

tokenizer = AutoTokenizer.from_pretrained("mirajbhandari/llma_deep_production")

model = AutoModelForCausalLM.from_pretrained("mirajbhandari/llma_deep_production", quantization_config=quantization_config)
```

```python
from transformers import pipeline
```

**Question 1**

```python
# Run text generation pipeline with our next model
prompt = "Where are your shoes produced?"
pipe = pipeline(task="text-generation", model=model, tokenizer=tokenizer, max_length=100)
result = pipe(f"[INST] {prompt} [/INST]")
print(result[0]['generated_text'])
```

**Trick to get short ans**

```python
prompt = "Can I purchase Deep Production shoes using cryptocurrency?"
pipe = pipeline(task="text-generation", model=model, tokenizer=tokenizer, max_length=300 )
result = pipe(f"[INST] {prompt} [/INST]")

generated_text = result[0]['generated_text']
print("this is how model gives other texts also: ",generated_text,"\n")

generated_sentences = generated_text.split(".")
answer = generated_sentences[0].strip()
print("this is how we make  model give one line ans : ",answer)

```
