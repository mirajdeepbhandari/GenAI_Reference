# Using Gpu

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("mirajbhandari/gemma-2b-it_yungri_final")
model = AutoModelForCausalLM.from_pretrained("mirajbhandari/gemma-2b-it_yungri_final", device_map="auto")
```

```python
# User input question
user_question = "give me contact and email of the yungri?"

que =[{"role": "user", "content": user_question}]

formatted_question=tokenizer.apply_chat_template(que, tokenize=False)

# Send formatted question to the model for prediction
device = "cuda:0"

inputs = tokenizer.encode(formatted_question , add_special_tokens=False, return_tensors="pt")

outputs = model.generate(input_ids=inputs.to(model.device), max_new_tokens=150)
conversation=tokenizer.decode(outputs[0],skip_special_tokens=True)

que=conversation.split("model")[0][4:].replace("\n","")
ans=conversation.split("model")[1].replace("\n","")
```

```python
print(f"Question:{que}\n\nAnswer:{ans}")
```

# Quantized Versions through bitsandbytes [ Using 8-bit precision (int8) ]

```python

from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

quantization_config = BitsAndBytesConfig(load_in_8bit=True)

tokenizer = AutoTokenizer.from_pretrained("mirajbhandari/gemma-2b-it_yungri_final")
model = AutoModelForCausalLM.from_pretrained("mirajbhandari/gemma-2b-it_yungri_final", quantization_config=quantization_config)

user_question = "how can i order tshirt in yungri?"

que =[{"role": "user", "content": user_question}]

formatted_question=tokenizer.apply_chat_template(que, tokenize=False)

inputs = tokenizer.encode(formatted_question , add_special_tokens=False, return_tensors="pt").to("cuda")

outputs = model.generate(input_ids=inputs, max_new_tokens=150)

conversation=tokenizer.decode(outputs[0],skip_special_tokens=True)

que=conversation.split("model")[0][4:].replace("\n","")

ans=conversation.split("model")[1].replace("\n","")

```

```python
print(f"Question:{que}\n\nAnswer:{ans}")
```
